```python
data['discount_price'] = (data['list_price']*data['discount_percent']) / 100
data.head(10)
```

```python
data['sale_price'] = data['list_price'] - data['discount_price']
data.head()
```

```python
data['profit'] = data['sale_price'] - data['cost_price']
data.head()
```

##### Find Top 10 Highest Revenue Generating Products

```python
query = """SELECT product_id, SUM(sale_price) AS sales from orders_data
            GROUP BY product_id
            ORDER BY sales DESC
            LIMIT 10"""
pd.read_sql(query, con)
```

##### Find Top 5 Highest Selling Products In Each Region

```python
query = """WITH cte AS (SELECT region, product_id, SUM(sale_price) as sales 
            FROM orders_data
            GROUP BY region, product_id)
            SELECT * FROM
            (SELECT *, row_number() over(partition by region order by sales desc) AS rank from cte)
            WHERE rank<=5"""

pd.read_sql(query, con)
```

##### Find Month Over Month Comparision for 2022 & 2023 Sales. Eg: Jan 2022 VS Jan 2023

```python
query = """WITH cte AS (
            SELECT CAST(strftime('%Y', order_date) AS INTEGER) AS order_year, 
            CAST(strftime('%m', order_date) AS INTEGER) AS order_month, 
            SUM(sale_price) as sales
            from orders_data
            GROUP BY order_year, order_month
            )
            SELECT order_month,
            SUM(CASE WHEN order_year = 2022 THEN sales ELSE 0 END) AS sales_2022,
            SUM(CASE WHEN order_year = 2023 THEN sales ELSE 0 END) AS sales_2023
            FROM cte
            GROUP BY order_month
            ORDER BY order_month
        """

pd.read_sql(query, con)
```

##### For Each Category Wich Month Had Highest Sales

```python
query = """WITH cte AS(
            SELECT category, strftime('%Y-%m', order_date) AS order_year_month, SUM(sale_price) as sales
            from orders_data
            GROUP BY category, order_year_month
            )
            SELECT * FROM (
            SELECT *, row_number() over(partition by category order by sales desc) AS rank
            from cte
            )
            where rank == 1
        """

pd.read_sql(query, con)
```

##### Which Sub-category Had Highest Growth By Profit In 2023 Compare To 2024

```python
query = """WITH cte AS (
            SELECT sub_category,
            CAST(strftime('%Y', order_date) AS INTEGER) AS order_year, 
            SUM(sale_price) AS sales
            FROM orders_data
            GROUP BY sub_category, order_year
            )
            , cte2 AS (
            SELECT sub_category,
            SUM(CASE WHEN order_year = 2022 THEN sales else 0 end) AS sales_2022,
            SUM(CASE WHEN order_year = 2023 THEN sales else 0 end) AS sales_2023
            from cte
            GROUP BY sub_category
            )
            SELECT *,
            ((sales_2023-sales_2022)*100/sales_2022) AS growth
            from cte2
            ORDER BY growth DESC
            LIMIT 1
        """

pd.read_sql(query, con)
```
